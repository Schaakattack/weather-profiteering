```python
import os
import glob
import pandas as pd
from pathlib import Path
import hvplot.pandas
import matplotlib.pyplot as plt
```

```python
# Flight data
path = r"./Airline_Data/"
all_files = glob.glob(os.path.join(path, "*arrival*.csv"))

data = []

for file in all_files:
    df = pd.read_csv(file, index_col='date', parse_dates=True, infer_datetime_format=True)
    df.groupby('date').sum()
    data.append(df)

arrival_df = pd.concat(data, ignore_index=True)

arrival_df[:-2]

arrival_df.to_csv('./aggregate_data/aggregate_arrivals_data.csv', index=False)

arrival_df
```

```python
# Stock price
path = r"./Airline_Data/"
all_files = glob.glob(os.path.join(path, "*stock*.csv"))

data = []

for file in all_files:
    df = pd.read_csv(file, index_col='date', parse_dates=True, infer_datetime_format=True)
    dl_stock_return = df.pct_change().dropna()
    dl_stock_pct = dl_stock_return.drop(columns=['open',
                                                 'high',
                                                 'low',
                                                 'close',
                                                 'volume'])
    data.append(dl_stock_pct)

stock_df = pd.concat(data, axis=1)

cols = ['AA', 'AAL', 'ALK', 'DE']

stock_df.columns = cols

arrival_df.to_csv('./aggregate_data/aggregate_stock_data.csv', index=False)

#stock_df.head(50)
```
