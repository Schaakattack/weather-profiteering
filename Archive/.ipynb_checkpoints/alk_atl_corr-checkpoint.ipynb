```python
# Import libraries an drequired dependencies
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Import the Alaska stock prices 2019 from the Airline Data
alk_df = pd.read_csv(
    Path("./Airline_Data/ALK_stock_2019.csv"),
    index_col='date',
    parse_dates=True,
    infer_datetime_format=True)
alk_df.dtypes
```

```python
# Read the first five rows of the data
alk_df.head()
```

```python
# Loc in columns required for analysis
alk_adj_close_df = alk_df.loc[:, [                                     
    'adj_close',
]]
alk_adj_close_df.head()
```

```python
# Review statistical data of DataFrame
alk_adj_close_df.describe()
```

```python
# Import Atlanta's's weather data
atl_df = pd.read_csv(
    Path("./Airline_data/atl_weather_2019.csv"),
    index_col='datetime',
    parse_dates=True,
    infer_datetime_format=True)
atl_df.dtypes
```

```python
# Review the first five rows of data
atl_df.head()
```

```python
# Loc in columns required for analysis
atl_analysis_df = atl_df.loc[:, ['precip',
    'windspeed',
    'visibility'   
]]
    
atl_analysis_df.head()
```

```python
# Create corelation between the 'adj_close' column of stock data 
# and 'precip' of weather data.
alk_adj_close_df['adj_close'].corr(atl_analysis_df['precip'])
```

```python
# Create corelation between the 'adj_close' column of stock data 
# and 'precip' of weather data.
alk_adj_close_df['adj_close'].corr(atl_analysis_df['windspeed'])
```

```python
# Create corelation between the 'adj_close' column of stock data 
# and 'precip' of weather data.
alk_adj_close_df['adj_close'].corr(atl_analysis_df['visibility'])
```
